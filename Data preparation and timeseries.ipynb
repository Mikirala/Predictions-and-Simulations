```python
del df2['Sensor_ID']
del df2['Sensor_Name']
del df2['ID']
del df2['Date_Time']
```

```python
df2.sort_index(inplace=True)
```

```python
sns.lineplot(data= df2.Hourly_Counts)
#shows regime change when COVID appeard
```

```python
df2.head()
```

```python
df3 = df2.loc[df2['Year'] < 2020]
df3.tail()
```

```python
df3['Trend'] = range(1,len(df3)+1)
df3.head()
```

```python
df3['LogPed'] = np.log()
```

```python
import statsmodels.formula.api as sm

model = sm.ols(formula='Hourly_Counts ~ Trend + C(Day) + C(Month) + C(Time)' , data=df3)
results = model.fit()
results.summary()
```

```python
results.fittedvalues
```

```python
df3['FinalReg'] = results.predict()
```

```python
sns.lineplot(data=df3.filter(['Hourly_Count','FinalReg']))
```
