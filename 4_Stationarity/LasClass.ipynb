```python
df = pd.read_csv(path + 'CL=F.csv')
df.index = df['Unnamed: 0'].astype('datetime64')
del df['Unnamed: 0']
df = df.filter(items=['CL=F.Adjusted'])
df
```

```python
df2 = pd.read_csv(path + 'GC=F.csv')
df2.index = df2['Unnamed: 0'].astype('datetime64')
del df2['Unnamed: 0']
df2 = df2.filter(items=['GC=F.Adjusted'])
df2
```

```python
df3 = pd.read_csv(path + 'GSPC.csv')
df3.index = df3['Unnamed: 0'].astype('datetime64')
del df3['Unnamed: 0']
df3 = df3.filter(items = ['GSPC.Adjusted'])
df3
```

```python
final = pd.merge(df, df1, left_index=True, right_index=True)
final = pd.merge(final, df2, left_index=True, right_index=True)
final = pd.merge(final, df3, left_index=True, right_index=True)
final = pd.merge(final, df4, left_index=True, right_index=True)
final = pd.merge(final, df5, left_index=True, right_index=True)
final

```

```python
import statsmodels.api as sm
sm.tsa.acf(final['CL=F.Adjusted'])
from statsmodels.graphics import tsaplots
import matplotlib.pyplot as plt
#plot autocorrelation function
fig = tsaplots.plot_acf(final['CL=F.Adjusted'], lags=10)
plt.show()
```
